```python
from src.noise import OUNoise, LinearSegmentNoise, NormalNoise, SmoothNoiseND

%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
```

```python
SIGMA = 0.15
THETA = 0.2
STEPS = 200
```

```python
def plot_orb(noise_process):
    orb = []

    for i in range(STEPS):
        orb.append(noise_process())

    ax = plt.axes(projection='3d')
    orb = np.array(orb)
    orb.shape
    ax.plot3D(orb[:,0], orb[:,1], orb[:,2], 'red')
```

```python
ou_noise = OUNoise(
    dim=3,
    sigma=SIGMA,
    theta=THETA,
    dt=1e-2)

plot_orb(ou_noise)
```

```python
normal_noise = NormalNoise(
    dim=3,
    sigma=0.2)

plot_orb(normal_noise)
```

```python
# Problem here is that the prob_value is an indirect bound on the nosie amplitude. So Sigma bec 

linear_segment_noise = LinearSegmentNoise(
    dim=3,
    sigma=0.2)

plot_orb(linear_segment_noise)
```

```python
smooth_noise = SmoothNoiseND(
    steps=STEPS,
    dim=3,
    sigma=0.2)

plot_orb(smooth_noise)
```
